```python
import transformers
import gradio as gr
import PIL
import IPython
import base64
import dotenv
import io
```

```python
img_cap = transformers.pipeline("image-to-text", model= "Salesforce/blip-image-captioning-large")
```

```python
image_url = "https://www.petful.com/wp-content/uploads/2015/05/Turkish-Van-cat-1-750x507.jpg"
IPython.display.display(IPython.display.Image(url= image_url)
)
```

```python
def captioner(image):
    caption = img_cap(image)
    return caption[0]["generated_text"]
```

```python
gr.close_all()
cap_app = gr.Interface(fn= captioner,
    inputs= [gr.Image(label= "upload image", type= "pil")],
    outputs= [gr.Textbox(label= "caption")],
    title="Image Captioning with BLIP",
    description="Caption any image using the BLIP model",
    allow_flagging="never",
)

cap_app.launch(share=True, debug= False)
```
